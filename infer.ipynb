```python
# 定义 PSNR 计算函数
import numpy as np
import math
import cv2
from skimage.metrics import peak_signal_noise_ratio as psnr


# 读取图像
OUTPUT_PATH = "outputv3/4800/"
LABEL_PATH = "data/test/clean/"


print("PSNR for each image:")
psnr1 = psnr(cv2.imread(LABEL_PATH + "43.png"), cv2.imread(OUTPUT_PATH + "43.png"))
pnsr2 = psnr(cv2.imread(LABEL_PATH + "45.png"), cv2.imread(OUTPUT_PATH + "45.png"))
pnsr3 = psnr(cv2.imread(LABEL_PATH + "44.png"), cv2.imread(OUTPUT_PATH + "44.png"))
print('43.png' + " PSNR: ", psnr1)
print('45.png' + " PSNR: ", pnsr2)
print('44.png' + " PSNR: ", pnsr3)
print("Average PSNR: ", (psnr1 + pnsr2 + pnsr3) / 3)
```
